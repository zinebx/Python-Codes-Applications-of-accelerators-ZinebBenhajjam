```python
import math

def get_rigidity(kinetic_energy, rest_mass, charge):
    """
    Calculate magnetic rigidity.

    Parameters:
    - kinetic_energy: Kinetic energy of the charged particle (in joules).
    - rest_mass: Rest mass of the charged particle (in kilograms).
    - charge: Charge of the particle (in coulombs).

    Returns:
    - Magnetic rigidity (R) in Tesla-meter (Tm).
    """
    # Convert kinetic energy to momentum
    momentum = math.sqrt(2 * rest_mass * kinetic_energy + kinetic_energy**2)

    try:
        rigidity = momentum / abs(charge)
        return rigidity
    except ZeroDivisionError:
        print("Error: Division by zero. Please provide a non-zero charge.")


```
